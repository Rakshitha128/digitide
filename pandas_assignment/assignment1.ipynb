```python
shipment["Cost_per_Mile"] = shipment["Cost"] / shipment["Distance_miles"]
shipment["Cost_per_Mile"]
```

```python
shipment["Late_Delivery"] = shipment["Status"] == "Delayed"
shipment["Late_Delivery"]
```

```python
import numpy as np
shipment["Weight_Normalized"] = (shipment["Shipment_Weight_kg"] - shipment["Shipment_Weight_kg"].min()) / \
                                (shipment["Shipment_Weight_kg"].max() - shipment["Shipment_Weight_kg"].min())
shipment["Weight_Normalized"]
```
